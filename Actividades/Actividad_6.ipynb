## **Primer paso: importar librerias**

```python
# Librerías
import matplotlib.pyplot as plt # generar gráficos
import numpy as np # generar arreglos
import pandas as pd # cargar conjuntos de datos externos
import sklearn.linear_model # contiene diferentes modelos de ML
from sklearn.datasets import fetch_california_housing

california_housing = fetch_california_housing(as_frame=True)
```

## **Segundo paso: importar conjunto de datos**

```python
california = fetch_california_housing()

df = pd.DataFrame(data=california.data, columns=california.feature_names)

df['Precio'] = california.target
```

## **Tercer paso: procesamiento de los pasos**

```python
X = df[['MedInc']] # MedInc (Ingresos medios en la zona)

y = df['Precio']
```

```python
df
```

```python
 df.plot(kind="scatter", x="MedInc", y="Precio")
plt.show()
```

## **Cuarto paso: seleccionar el modelo de aprendizaje y entrenarlo**

```python
model = sklearn.linear_model.LinearRegression()
```

```python
model.fit(X,y)
```

```python
X
```

# **Quinto paso: poner a prueba el modelo entrenado**

```python
X_new= [[4]]
print(model.predict(X_new))
```

```python
plt.scatter(X, y, color='red')
plt.xlabel("Medinc")
plt.ylabel("Precio")
plt.title("Mi primer modelo de regresión lineal:b")
plt.ylim(0, 5)
plt.plot(X, model.predict(X), color='blue')
```
