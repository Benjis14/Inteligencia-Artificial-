# **PASO 1 : CARGAR LIBRERIAS**

```python
# TensorFlow y tf.keras
import tensorflow as tf
from tensorflow import keras

# Librerias de ayuda
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)
```

# **PASO 2: CARGAR DATOS**

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
class_names = ['cero', 'uno', 'dos', 'tres', 'Cuatro',
               'cinco', 'Seis', 'Siete', 'ocho', 'nueve']
```

```python
x_train = x_train / 255.0

x_test = x_test / 255.0
```

```python
plt.figure(figsize=(10, 10))  # Establece el tamaño de la figura
for i in range(25):
    plt.subplot(5, 5, i + 1)  # Crea una cuadrícula de 5x5 para mostrar las imágenes
    plt.xticks([])  # Elimina las marcas del eje x
    plt.yticks([])  # Elimina las marcas del eje y
    plt.grid(False)  # Desactiva la cuadrícula
    plt.imshow(x_train[i], cmap=plt.cm.binary)  # Muestra la imagen en escala de grises
    plt.xlabel(class_names[y_train[i]])  # Etiqueta la imagen con su clase correspondiente

plt.show()  # Muestra todas las imágenes
```

# **PASO 4: SELECCIÓN Y ENTRENAMIENTO DEL MODELO**

```python
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Input(shape=(28, 28)),  # Usando Input en lugar de input_shape
    keras.layers.Flatten(),                # Aplana la entrada
    keras.layers.Dense(128, activation='relu'),  # Capa densa con 128 neuronas
    keras.layers.Dense(10, activation='softmax')  # Capa de salida con 10 neuronas
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10)
```

```python
# Evaluación del modelo en el conjunto de prueba
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)

# Mostrar la precisión en el conjunto de prueba
print('\nPrecisión en el conjunto de prueba:', test_acc)
```

# **PASO 5: HACER PREDICCIONES**

```python
# Realiza predicciones en el conjunto de prueba
predictions = model.predict(x_test)
```

```python
def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array, true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  predictions_array, true_label = predictions_array, true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')
```

```python
# Asegúrate de haber definido las funciones plot_image y plot_value_array previamente

i = 0  # Cambia este índice para ver otras imágenes
plt.figure(figsize=(6, 3))

# Subgráfico para la imagen
plt.subplot(1, 2, 1)
plot_image(i, predictions[i], y_test, x_test)  # Cambié test_labels por y_test

# Subgráfico para el gráfico de barras de predicción
plt.subplot(1, 2, 2)
plot_value_array(i, predictions[i], y_test)

# Mostrar la figura
plt.show()
```

```python
i = 12  # Cambiar el índice para visualizar la imagen en la posición 12
plt.figure(figsize=(6, 3))  # Tamaño de la figura

# Subgráfico para mostrar la imagen
plt.subplot(1, 2, 1)
plot_image(i, predictions[i], y_test, x_test)  # Visualiza la imagen y la etiqueta verdadera

# Subgráfico para mostrar el gráfico de barras de predicciones
plt.subplot(1, 2, 2)
plot_value_array(i, predictions[i], y_test)  # Visualiza las probabilidades de las predicciones

# Mostrar la figura
plt.show()
```

```python
# Plot the first X test images, their predicted labels, and the true labels.
# Color correct predictions in blue and incorrect predictions in red.
num_rows = 5
num_cols = 3
num_images = num_rows * num_cols
plt.figure(figsize=(2 * 2 * num_cols, 2 * num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2 * num_cols, 2 * i + 1)
    plot_image(i, predictions[i], y_test, x_test)  # Cambiado a y_test y x_test
    plt.subplot(num_rows, 2 * num_cols, 2 * i + 2)
    plot_value_array(i, predictions[i], y_test)  # Cambiado a y_test
plt.tight_layout()
plt.show()
```
